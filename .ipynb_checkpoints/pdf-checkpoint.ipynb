```python
import numpy as np
import pandas as pd
from tabula.io import read_pdf
```

```python
df=tabula.io.read_pdf('../data/379100000Grest09801.pdf',pandas_options={'header': None},pages='all')
```

```python
# 查有多少DataFrame
len(df)
```

```python
# 給予欄位名稱
df[0].columns=["序號","統一編號","商號名稱","營業所在地","資本額組/織型態","負責人","變更日期","核准文號"]
```

```python
df1=df[0]
```

```python
df1
```

```python
# 被合併的欄位拆分，給予新欄位
df1['資本額']=df1['資本額組/織型態'].str.split(" ").str.get(0)

df1['組織型態']=df1['資本額組/織型態'].str.split(" ").str.get(1)
```

```python
# 除去不需要的欄位
df1.drop(['資本額組/織型態','序號','負責人','核准文號'],axis=1,inplace=True)
```
