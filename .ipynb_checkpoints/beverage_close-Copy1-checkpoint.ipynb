```python
import numpy as np
import pandas as pd
```

```python
# 飲料業登記整理
close1310_2204 = pd.read_csv('../data/close1310_2204.csv')

# 與登記合併過的
regist = pd.read_csv('../data/regist1310_2204.csv')
```

```python
regist.info()
```

```python
# 合併兩資料取交集
test = pd.merge(close1310_2204,regist,on='統一編號')
test.info()
```

```python
test.head(20)
```

```python
beverage=pd.read_csv('../career_raw/飲料店業.csv')
```

```python
beverage=pd.merge(beverage,test,on='統一編號')
```

```python
beverage.info()
```

```python
beverage.sort_values(by=['設立日期'],inplace=True)
```

```python
beverage.tail(20)
```

```python
# 確認欄位相同
beverage_dif=beverage.query('商業名稱_x != 商業名稱_y | 商業所在地_y != 商業所在地_x')
beverage_dif.info()
```

```python
# 欄位相同
beverage=beverage.query('商業名稱_x == 商業名稱_y & 商業所在地_x == 商業所在地_y')
# 去除重複欄位
beverage.drop(['商業名稱_y','商業所在地_x'],axis=1,inplace=True)
```

```python
beverage.drop(['縣市名稱_y','組織型態_y','資本額_y','商業名稱_x','商業所在地_y'],axis=1,inplace=True)
beverage.info()
```

```python
beverage['歇業日期']=pd.to_datetime(beverage['歇業日期'],format='%Y%m')
```

```python
beverage['設立日期']=pd.to_datetime(beverage['設立日期'],format='%Y%m')
```

```python
beverage.info()
```

```python
beverage['開業長度']=(beverage['歇業日期']-beverage['設立日期'])
```

```python
change=pd.read_csv('../data/change1310_2204.csv')
```
